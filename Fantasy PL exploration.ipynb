```python
def plotposition(position):
    points = [points for points, _ in positions[position]]
    minutes = [minutes for _, minutes in positions[position]]
    
    LX = np.arange(0, 90)
    m, b, r, p, err = scipy.stats.linregress(minutes, points)
    plt.scatter(minutes, points)
    plt.plot(LX, m * LX + b, label='Best Fit')
    plt.xlabel("minutes")
    plt.ylabel("points")

plotposition("Forward")
```

```python
opponents = {}
for player in players.itervalues():
    for game in player["fixture_history"]["all"]:
        #skip games where the player played 0 minutes
        if game[3] == 0: continue
        opp = game[2][:3]
        pts = game[19]
        opponents.setdefault(opp, [0,0])[0] += pts
        opponents[opp][1] += 1

from collections import OrderedDict
avgs = {}
for opponent, (score, n) in opponents.iteritems():
    avgs[opponent] = score/float(n)

sorted_avgs = OrderedDict(sorted(avgs.items(), key=lambda t: t[1]))
```

```python
#wow that's a big difference! A game against crystal palace is 2x more valuable than one against Man City!
class Game(object):
    def __init__(self, game_json):
        self.opp = game_json[2][:3]
        # "A" for away, "H" for home
        self.loc = game_json[2][4]
        self.points = game_json[19]
        self.minutes = game_json[3]
    
    def __repr__(self):
        return "Game vs. %s %s: %s pts" % (self.opp, self.loc, self.points)

team_abbreviations = {
    'Norwich': 'NOR',
    'Cardiff City': 'CAR',
    'Man City': 'MCI',
    'Newcastle': 'NEW',
    'West Brom': 'WBA',
    'West Ham': 'WHU',
    'Southampton': 'SOU',
    'Sunderland': 'SUN',
    'Stoke City': 'STK',
    'Stoke': 'STK',
    'Crystal Palace': 'CRY',
    'Arsenal': 'ARS',
    'Swansea': 'SWA',
    'Liverpool': 'LIV',
    'Hull City': 'HUL',
    'Man Utd': 'MUN',
    'Everton': 'EVE',
    'Fulham': 'FUL',
    'Tottenham': 'TOT',
    'Spurs': 'TOT',
    'Aston Villa': 'AVL',
    'Chelsea': 'CHE',
    'Burnley': 'BUR',
    'QPR': 'QPR',
    'Leicester': 'LEI',
    'Hull': 'HUL',
}
    
class Player(object):
    def __init__(self, player_json):
        self.games = [Game(g) for g in player_json["fixture_history"]["all"]]
        self.name = u"{first_name} {second_name}".format(**player_json)
        self.cost = player_json["now_cost"]
        self.position = player_json["type_name"]
        self.pos = self.shortname(self.position)
        self.upcoming = self.get_upcoming_fixtures(player_json["fixtures"]["all"])
        self.raw = player_json
        self.team = team_abbreviations[player_json["team_name"]]
        self.idn = player_json["id"]
        self.news = player_json["news"]

    def get_upcoming_fixtures(self, fixtures):
        upcoming = []
        for _, gameweek, opponent in fixtures:
            week = int(gameweek.split()[-1])
            opp, loc = opponent.split('(')
            opp = team_abbreviations[opp.strip()]
            loc = loc[0]
            upcoming.append((week, opp, loc))
        return upcoming
    
    def shortname(self, position):
        pos_abbreviations = {
            "Goalkeeper": "gk",
            "Defender": "d",
            "Midfielder": "m",
            "Forward": "f"
        }
        
        return pos_abbreviations[position]

    def points(self):
        return sum(g.points for g in self.games)
    
    def last_n_points(self, n=5):
        return sum(g.points for g in self.games[-n:])
    
    def __repr__(self):
        return "#%s %s %s $%s %s %s" % (self.idn, self.team, self.name.encode("utf8"), self.cost, self.pos, self.points())

    def __unicode__(self):
        return "#%s %s $%s %s p%s" % (self.idn, self.name, self.cost, self.pos, self.points())

        return u"Player: %s %s" % (self.name, self.cost)

player_objs = [Player(p) for p in players.itervalues()]
print player_objs[12].points()
print player_objs[587]
#print player_objs[1].upcoming
#print player_objs[1].raw
```

```python
# measure home/away difference

# let's just consider players who played 90 minutes to eliminate minute bias
away = 0.
awayn = 0.
home = 0.
homen = 0.
for player in player_objs:
    for game in player.games:
        if game.minutes == 90:
            if game.loc == "A":
                away += game.points
                awayn += 1
            else:
                home += game.points
                homen += 1
                
print away, awayn, home, homen, away/awayn, home/homen

homefield_advantage = home/homen - away/awayn

print "Home Field Advantage is worth about %.2f points" % homefield_advantage
```

```python
# now what I want is a function to guess a player's average score on a neutral pitch against average opponents
avg_opponent = sum(avgs.values())/float(len(avgs))
homefield_adjustment = homefield_advantage/2
def adjusted_score(game):
    pts = game.points
    pts += homefield_adjustment if game.loc == "A" else -homefield_adjustment
    pts += avg_opponent - avgs[game.opp]
    return pts

def player_avg(player):
    return sum(g.points for g in player.games) / float(len(player.games))

def adjusted_player_avg(player, n=5):
    return sum(adjusted_score(g) for g in player.games[-n:]) / n

print player_objs[2].games
print adjusted_score(player_objs[2].games[2])
print adjusted_player_avg(player_objs[2]), player_avg(player_objs[2])

print player_objs[12].games
print adjusted_player_avg(player_objs[12]), player_avg(player_objs[12])
```

```python
#now what I want is a function to predict the player's scores in the next n gameweeks
def expected_scores(player, n=5):
    thisweek = player.upcoming[0][0]
    # make sure that we count double gameweeks
    fixtures = [g for g in player.upcoming if g[0] < thisweek + n]
    avg = adjusted_player_avg(player)
    
    expected = []
    for week, opp, loc in fixtures:
        pts = avg
        pts += homefield_adjustment if loc == "H" else -homefield_adjustment
        pts -= avg_opponent - avgs[opp]
        #print adjusted_player_avg(player), pts, opp, loc, week, avg_opponent-avgs[opp], homefield_adjustment if loc == "H" else -homefield_adjustment

        expected.append(pts)
    
    return sum(expected)/n

print unicode(player_objs[95]), player_objs[95].games
expected_scores(player_objs[95])
```

```python
import subprocess, re
def get_id(idn):
    for p in player_objs:
        if p.idn == idn: return p
    raise ValueError("Unable to find player")
    
def return_team(lp):
    val = subprocess.check_output("echo '%s' | lp_solve" % lp, shell=True).split('\n')
    team_ids = [int(re.search("^\w+?(\d+)", l).group(1)) for l in val if re.search(r" 1$", l)]
    return map(lambda x: get_id(int(x)), team_ids)

#return_team(out.getvalue())
```

```python
import StringIO
out = StringIO.StringIO()

#let's make an lp_solve just for gks, to test it out

# http://lpsolve.sourceforge.net/5.5/lp-format.htm
# http://craven-projects.blogspot.com/2010/08/choosing-fantasy-football-team-using.html

gks = [p for p in player_objs if p.position == "Goalkeeper"]

gk_score_constraints = []
for player in gks:
    pts = expected_scores(player)
    gk_score_constraints.append("{pts} gk{player.idn}".format(**locals()))

out.write("max: " + " + ".join(gk_score_constraints) + ";\n")

#gk_price_contraints = []
#for player in gks:
prices = " + ".join(("{p.cost} gk{p.idn}".format(**locals()) for p in gks))
out.write("price_limit: " + prices + " <= 100;\n")

gk_list = " + ".join(("gk{p.idn}".format(**locals()) for p in gks))
out.write("player_limit: " + gk_list + " = 2;\n")

variables = ", ".join("gk{p.idn}".format(**locals()) for p in gks)
out.write("bin %s;\n" % variables)
    
return_team(out.getvalue())
```

```python
#it works! Sweeto! Now let's try it for everyone!
out = StringIO.StringIO()

def objective_function():
    score_constraints = []
    for p in player_objs:
        pts = expected_scores(p)
        score_constraints.append("{pts} {p.pos}{p.idn}".format(**locals()))
    
    return "max: " + " + ".join(score_constraints) + ";\n"

out.write(objective_function())

def cost_constraint(max_price):
    cost_constraints = []
    for p in player_objs:
        cost_constraints.append(("{p.cost} {p.pos}{p.idn}".format(**locals())))
    costs = " + ".join(cost_constraints)
    
    return "cost_constraint: " + costs + " <= %s;\n" % max_price

out.write(cost_constraint(1002))

def position_constraints():
    constraints = StringIO.StringIO()

    gks = [p for p in player_objs if p.position == "Goalkeeper"]
    gk_list = " + ".join(("gk{p.idn}".format(**locals()) for p in gks))
    constraints.write("gk_limit: " + gk_list + " = 2;\n")
    
    ds = [p for p in player_objs if p.position == "Defender"]
    d_list = " + ".join(("d{p.idn}".format(**locals()) for p in ds))
    constraints.write("d_limit: " + d_list + " = 5;\n")
    
    ms = [p for p in player_objs if p.position == "Midfielder"]
    m_list = " + ".join(("m{p.idn}".format(**locals()) for p in ms))
    constraints.write("m_limit: " + m_list + " = 5;\n")
    
    fs = [p for p in player_objs if p.position == "Forward"]
    f_list = " + ".join(("f{p.idn}".format(**locals()) for p in fs))
    constraints.write("f_limit: " + f_list + " = 3;\n")
    
    return constraints.getvalue()

out.write(position_constraints())

def all_player_variables():
    variables = ", ".join("{p.pos}{p.idn}".format(**locals()) for p in player_objs)
    return "bin %s;\n" % variables

out.write(all_player_variables())
    
return_team(out.getvalue())
```

```python
#I'm in heaven! But too much Arsenal... let's add a team constraint.

def team_constraints(n=2):
    teams = {}
    team_constraints = StringIO.StringIO()
    for p in player_objs:
        teams.setdefault(p.team, []).append(p)
    
    for team, players in teams.iteritems():
        p_list = []
        for p in players:
            p_list.append("{p.pos}{p.idn}".format(p=p))
        team_constraints.write("{team}: ".format(team=team) + " + ".join(p_list) + " <= 2;\n")
    
    return team_constraints.getvalue()

out = StringIO.StringIO()
out.write(objective_function())
out.write(cost_constraint(1003))
out.write(position_constraints())
out.write(team_constraints())
out.write(all_player_variables())

#print re.findall("\d+ m13 ", out.getvalue())
    
team = return_team(out.getvalue())
print sum(p.cost for p in team)
team
```
